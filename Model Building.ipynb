## Model Building

##### Linear regression

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
print("Accuracy of the Linear Regression Model on Training set is : {}% and on Test set is {}%".format(round(lr_model.score(X_train, y_train),4)*100, round(lr_model.score(X_test, y_test),4)*100))
```

##### LGBM Regressor

```python
from lightgbm import LGBMRegressor
lc = LGBMRegressor(objective ="regression", boosting_type='gbdt', num_leaves=50, min_child_samples=19, subsample_freq=1, bagging_fraction=0.9, n_estimators=1000, max_depth=200, learning_rate=0.1, random_state=42)
lc.fit(X_train, y_train)
print("Accuracy of the LGBM Regression Model on Training set is : {}% and on Test set is {}%".format(round(lc.score(X_train, y_train),4)*100, round(lc.score(X_test, y_test),4)*100))
```

##### XGBoost

```python
import xgboost as xgb
xgb_model = xgb.XGBRegressor(n_estimators=120, reg_alpha=1e-5,subsample=0.8, base_score=0.8 , gamma=0.9 ,colsample_bylevel=0.1, learning_rate=0.1, max_depth=120, random_state=42)
xgb_model.fit(X_train, y_train)
print("Accuracy of the XGBoost Regression Model on Training set is : {}% and on Test set is {}%".format(round(xgb_model.score(X_train, y_train),4)*100, round(xgb_model.score(X_test, y_test),4)*100))
```

##### CatBoost

```python
import catboost as cb
cb_model = cb.CatBoostRegressor(iterations=1009,rsm=0.2,
                           learning_rate=1,
                           depth=3)
cb_model.fit(X_train, y_train)
print("Accuracy of the catBoost Regression Model on Training set is : {}% and on Test set is {}%".format(round(cb_model.score(X_train, y_train),4)*100, round(cb_model.score(X_test, y_test),4)*100))
```
